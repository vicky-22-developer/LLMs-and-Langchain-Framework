#q1> What are Large Language Models (LLMs) and how do they function?

ans -> An LLM is a type of model in Natural Language Processing (NLP) trained on huge amounts of text data (books, websites, articles, etc.).


 > how do they function?

1. Training on massive data
LLMs read billions of sentences.
They learn patterns like:
Grammar
Meaning
Context
Relationships between words

Example:

“I am going to the ___” → model predicts “market”, “school”, etc.

2. Based on Transformer architecture

Most modern LLMs use a neural network design called the
 Transformer architecture

This allows the model to:

Look at all words in a sentence at once
Understand context better than older models
3. Attention mechanism (key idea)

Inside Transformers is something called attention:

The model focuses on important words in a sentence
It understands relationships

Example:

“The dog chased the cat because it was fast”
 “it” refers to dog, not cat (model figures this out)

4. Tokenization

Before processing:

Text is broken into small units called tokens
Tokens can be:
words
parts of words
characters
5. Prediction (core functioning)

LLMs don’t “think” like humans.
They predict the next word/token step by step.

Example:

“India is a”
 predicts “country” with high probability

Then continues generating sentence.

6. Fine-tuning & alignment

After basic training:

Models are improved using:
human feedback
specific datasets
This helps them:
give better answers
avoid harmful responses

 #q5> LangChain and LlamaIndex are popular tools used in building LLM-based applications.
ans> LangChain focuses on managing the overall workflow of an application.
It helps in chaining multiple LLM calls and integrating APIs and tools.
It also supports memory and agents for decision-making tasks.
Thus, LangChain solves the problem of building complex, multi-step AI systems.
On the other hand, LlamaIndex focuses on connecting LLMs with external data.
It allows data from PDFs, documents, and databases to be indexed.
This makes it easier for LLMs to retrieve relevant information.
Thus, LlamaIndex solves the problem of using custom or private data.
LangChain is mainly used for application logic and orchestration.
LlamaIndex is mainly used for data retrieval and indexing.
LangChain controls how the system works, while LlamaIndex controls what data is used.
Both tools can be used together in real-world applications.
LlamaIndex provides data, and LangChain processes it.
Therefore, they complement each other in building efficient LLM systems.


```python
#q6>  Implement a basic Langchain pipeline using OpenAI’s LLM to answer questions based on a user input prompt.

!pip install langchain langchain-openai openai

from langchain_openai import OpenAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain

import os
os.environ["OPENAI_API_KEY"] = "sk-xxxxxxxxxxxxxxxx"
llm = OpenAI(temperature=0.7)
prompt = PromptTemplate(
    input_variables=["question"],
    template="Answer clearly:\n{question}"
)
chain = LLMChain(llm=llm, prompt=prompt)
question = "What is Python?"
response = chain.invoke({"question": question})
print("Answer:", response["text"])
```

```python
#q7> Integrate Langchain with a third-party API (e.g., weather, news) and show how responses can be generated via LLMs.


!pip install langchain langchain-openai requests

import os
import requests
from langchain_openai import ChatOpenAI
from langchain.prompts import PromptTemplate

os.environ["OPENAI_API_KEY"] = "your_openai_api_key"
WEATHER_API_KEY = "your_weather_api_key"

def get_weather(city):
    url = f"http://api.weatherapi.com/v1/current.json?key={WEATHER_API_KEY}&q={city}"
    response = requests.get(url)
    data = response.json()

    if "current" not in data:
        return "Weather data not found. Check API key or city name."

    temp = data["current"]["temp_c"]
    condition = data["current"]["condition"]["text"]

    return f"The current temperature in {city} is {temp}°C with {condition}."

llm = ChatOpenAI(model="gpt-3.5-turbo", temperature=0.7)
prompt = PromptTemplate.from_template(
    "Convert this weather data into a friendly response:\n{weather_info}"
)

chain = prompt | llm

city = input("Enter city name: ")

weather_info = get_weather(city)

response = chain.invoke({"weather_info": weather_info})

print(response.content)
```

```python
#q9>Demonstrate combining Langchain with LamaIndex to create a simple document-based Q&A chatbot.

!pip install llama-index langchain langchain-openai

import os
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader
from langchain_openai import ChatOpenAI

os.environ["OPENAI_API_KEY"] = "your_openai_api_key"

documents = SimpleDirectoryReader("data").load_data()

index = VectorStoreIndex.from_documents(documents)
query_engine = index.as_query_engine()

llm = ChatOpenAI(model="gpt-3.5-turbo", temperature=0.7)

print(" Document Q&A Chatbot (type 'exit' to stop)\n")

while True:
    user_query = input("You: ")

    if user_query.lower() == "exit":
        break

    response = query_engine.query(user_query)

    final_prompt = f"""
    Make this answer more conversational and clear:

    {response}
    """

    final_answer = llm.invoke(final_prompt)

    print("Bot:", final_answer.content)
```

 #q10>  A legal firm wants to use AI to summarize large volumes of legal
#documents and retrieve relevant information quickly. Propose a solution using
#Langchain and LamaIndex, and explain how it would work in practice.

ans> A legal firm can use LlamaIndex and LangChain to process large legal documents efficiently.
LlamaIndex is used to load, split, and index documents for fast retrieval.
When a user asks a query, it retrieves relevant sections of the document.
LangChain then processes this data using an LLM to generate summaries or answers.
This system helps in quickly analyzing contracts and extracting important clauses.
However, human verification is required for accuracy.

```python

!pip install llama-index langchain langchain-openai pypdf

import os
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader
from langchain_openai import ChatOpenAI

os.environ["OPENAI_API_KEY"] = "your_openai_api_key"

documents = SimpleDirectoryReader("data").load_data()

index = VectorStoreIndex.from_documents(documents)

query_engine = index.as_query_engine()

llm = ChatOpenAI(model="gpt-3.5-turbo", temperature=0.3)

print(" Legal Document Assistant (type 'exit' to stop)\n")

while True:
    user_query = input("You: ")

    if user_query.lower() == "exit":
        break

    retrieved_data = query_engine.query(user_query)

    prompt = f"""
    You are a legal assistant.
    Based on the following document data, answer clearly and highlight key legal points:

    {retrieved_data}

    Question: {user_query}
    """
    response = llm.invoke(prompt)

    print(" Answer:\n", response.content)
```
